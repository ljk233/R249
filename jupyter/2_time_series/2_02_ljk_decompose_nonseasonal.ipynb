## Summary

The aim of this notebook was to show how to decompose non-seasonal time series data using **R** so the trend component can be estimated.
Data on the average annual temperature in central England between 1659 and 2004 was plotted.
The series was found to be extremely noisy, making it difficult to identify any underlying trend.
Two simple moving averages were calculated and plotted, making use of the `SMA` function from the `TTR` package.
The differences in the plots of the moving averages were discussed.

## Get the data

Data on the average annual temperature in central England was plotted between 1659 and 2004 is shown below in a `data.frame`.

```r
annualtemps <- read.csv("..\\..\\data\\alltemps.csv")
head(annualtemps)
```

Declare a variable with a reference to the `temperature` column.

```r
alltemps <- annualtemps$temperature
```

## Plot the time series

Initialise a time series object using `alltemps` and plot the average annual annual temperature data.
The data is non-seasonal, given that it is comprised of annual averages, so do not pass an actual argument for the `frequency` parameter.

```r
tstemp <- ts(annualtemps$temperature, start = c(1659))
plot.ts(tstemp, xlab = "year", ylab = "temperature")
```

The time plot is very noisy, making it difficult to gauge if there is a trend in the time series.

## Simple moving averages

Minimise the irregular component by taking the the simple moving average of the series.
This is referred to as _smoothing_.
Two simple moving averages of different orders are plotted below, of orders 5 and 31.

```r
tstempSMA5 <- SMA(tstemp, n = 5)
plot.ts(tstempSMA5, xlab = "year", ylab = "ma5")
```

```r
tstempSMA31 <- SMA(tstemp, n = 31)
plot.ts(tstempSMA31, xlab = "year", ylab = "ma31")
```

The plot of **MA(5)** has reduced the noise compared to the original series, but it still not smooth enough to be certain in identifying a trend.
The noise is reduced even further in the plot of **MA(31)**, making the underlying the underlying trend much clearer.
Overall, it seems that **MA(31)** gives a better impression of the trend in the changing annual average temperature between 1659 and 2004.
