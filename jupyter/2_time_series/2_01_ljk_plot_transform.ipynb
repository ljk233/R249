## Get the data

```r
infections <- read.csv("..\\..\\data\\infections.csv")
head(infections)
```

```r
rota <- infections$rotavirus
```

## Initialise the time series

```r
tsrota <- ts(infections$rotavirus, start = c(1996, 27), frequency = 52)
```

## Plot the time series

```r
plot.ts(tsrota, xlab = "Week", ylab = "Weekly cases")
```

## Transform the time series

```r
plot.ts(log(tsrota), xlab = "Week", ylab = "log(Weekly cases)")
```
